```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Human trafficking 15-20.csv')

# Get the list of states
states = df.iloc[:, 0].unique()

# Calculate the total number of cases for each state
total_cases = []
for state in states:
    cases = df[df.iloc[:, 0] == state].iloc[:, 1:].squeeze().values.sum()
    total_cases.append(cases)

# Plot the total number of cases for each state
plt.figure(figsize=(12, 6))
plt.bar(states, total_cases)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total number of cases')
plt.title('Total number of cases for each state from 2015 to 2020')
plt.show()

# Calculate the number of cases per year for each state
cases_per_year = []
for state in states:
    cases = df[df.iloc[:, 0] == state].iloc[:, 1:].squeeze().values
    cases_per_year.append(cases)

# Plot the number of cases per year for each state
fig, axs = plt.subplots(len(states), figsize=(10, 60))
for i, state in enumerate(states):
    axs[i].plot(range(2015, 2021), cases_per_year[i])
    axs[i].set_xlabel('Year')
    axs[i].set_ylabel('Number of cases')
    axs[i].set_title(state)
plt.show()
```

```python
avg_cases_per_year = []
for state_cases in cases_per_year:
    avg_cases_per_year.append(state_cases.mean())

# Plot the average number of cases per year for each state
plt.figure(figsize=(12, 6))
plt.bar(states, avg_cases_per_year)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Average number of cases per year')
plt.title('Average number of cases per year for each state from 2015 to 2020')
plt.show()
```

```python
# Create a DataFrame of the number of cases by year and state
cases_df = pd.DataFrame(cases_per_year, columns=range(2015, 2021), index=states)

# Plot a heatmap of the number of cases by year and state
plt.figure(figsize=(10, 10))
plt.imshow(cases_df, cmap='YlOrRd')
plt.colorbar()
plt.xticks(range(len(cases_df.columns)), cases_df.columns, rotation=90)
plt.yticks(range(len(cases_df.index)), cases_df.index)
plt.xlabel('Year')
plt.ylabel('State')
plt.title('Number of cases by year and state from 2015 to 2020')
plt.show()

# Calculate the percentage change in the number of cases from 2015 to 2020 for each state
percent_change = []
for state_cases in cases_per_year:
    percent_change.append((state_cases[-1] - state_cases[0]) / state_cases[0] * 100)

# Plot the percentage change in the number of cases from 2015 to 2020 for each state
plt.figure(figsize=(12, 6))
plt.bar(states, percent_change)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Percentage change in the number of cases')
plt.title('Percentage change in the number of cases from 2015 to 2020 for each state')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Human trafficking 15-20.csv')

# Get the total number of cases for each state
total_cases = df.iloc[:, 1:].sum(axis=1)

# Create a new DataFrame with state and total cases
data = pd.DataFrame({'State': df.iloc[:, 0], 'Total Cases': total_cases})

# Sort the data by total cases
data_sorted = data.sort_values(by='Total Cases', ascending=False)

# Get the top 10 states with the highest number of cases
top_states = data_sorted.head(10)

# Get the top 10 states with the lowest number of cases
bottom_states = data_sorted.tail(10)

# Create a bar chart for the top 10 states with the highest number of cases
plt.figure(figsize=(10, 6))
plt.bar(top_states['State'], top_states['Total Cases'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total number of cases')
plt.title('Top 10 states with the highest number of cases over the entire time period')
plt.show()

# Create a bar chart for the top 10 states with the lowest number of cases
plt.figure(figsize=(10, 6))
plt.bar(bottom_states['State'], bottom_states['Total Cases'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total number of cases')
plt.title('Top 10 states with the lowest number of cases over the entire time period')
plt.show()
```
