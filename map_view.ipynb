```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load the geojson file
india_states = gpd.read_file('https://raw.githubusercontent.com/geohacker/india/master/district/india_district.geojson')

# Load the dataset
df = pd.read_csv('avg_cases.csv')

# Join the two datasets
merged_data = india_states.merge(df, left_on='NAME_1', right_on='State')

# Create a choropleth map
fig, ax = plt.subplots(1, figsize=(10, 6))
merged_data.plot(column='Cases', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.axis('off')
ax.set_title('Number of cases by state')
plt.show()
```
